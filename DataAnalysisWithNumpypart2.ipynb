```python
a = np.random.randint(1 ,10 , size =(2,3))
b = np.random.randint(20 ,30 , size =(2,3))
print(a)
print(b)
```

```python
con_row=np.concatenate((a,b),axis=0)
print(con_row)
con_col=np.concatenate((a,b),axis=1)
print(con_col)
```

```python
x=np.array([1,2,3,4,5,6])
```

```python
y=np.array([7,8,9,10,11,12])
sin_val = np.sin(y)
print(sin_val)
cos_val=np.cos(x)
print(cos_val)
log_2=np.log2(x)
print(log_2)
s=np.sum(x)
print(s)
```

```python
x = np.array([ 10 , 8 , 16 , 100 ])
z=x.copy()
print(z)
z.sort()

print(z)

print(x)
```

```python
# 2d array sorting
mat = np.array([[10,3,5]
                ,[8,4,9]])
hor_sort=np.sort(mat,axis=1)
print(hor_sort)
vertic_sort=np.sort(mat,axis=0)
print(vertic_sort)
```

```python
print(x)
index=np.where(x==100)
print(index)
arr=np.where(x>8,x,0)
print(arr)
```

```python
data=np.genfromtxt('student_scores.csv',delimiter=',',skip_header=1)
print(data)
```

```python
math_marks=data[::,:1]
# print(math_marks)
# print(math_marks.T)
max_marks=np.max(math_marks)
print(max_marks)
min_marks=np.min(math_marks)
print(min_marks)
median_marks=np.median(math_marks)
print(median_marks)
std_math_marks = np.std(math_marks)
print(std_math_marks)
```
