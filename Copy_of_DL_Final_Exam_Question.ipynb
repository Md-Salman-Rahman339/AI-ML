error is passed to the hidden layer using weights

Backward Flow of Error:

2.Moves backward layer by layer
3.Weights are updated contineously .
Flow Chart->Output Layer->Hidden Layer->Input Layer

# derive the mathematical expressions for weight updates using the chain rule:
Here d stands for del,so
dL/dw=dL/dy_pred * dy_pred/dz * dz/dw

1.dL/dy_pred=(y_pred-y)

2.dy_pred/dz=y_pred *(1-y_pred)
3.dz/dx=x

so,dL/dw=(y_pred-y) * y_pred *(1-y_pred) *x

so weight update rule:
w_new=w_old-learning rate * dL/dw
