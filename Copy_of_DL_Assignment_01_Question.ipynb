# What is meant by the Perceptron Learning Rule ?

The Perceptron Learning Rule is a method used to update the weights of  perceptron so  it can make correct predictions.When the perceptron makes a wrong prediction and weights are adjusted using the rule:
Wnew=Wold+learning rate*(y-y')*Xi​

weights are changed in a way that moves the decision boundary toward the correct class. This process is repeated until the perceptron classifies training data correctly.
