```python

import pandas as pd
import numpy as np

df = pd.read_csv('final-employee-ds.csv')
df
```

## Answer to the Q1:

```python
numpy_array = df['TrainingHours'].to_numpy()
# print(numpy_array)
float_numpy_array=numpy_array.astype(float)
print(float_numpy_array.dtype)
mn=np.mean(float_numpy_array)
dev=np.std(float_numpy_array)
print(mn,dev)

```

## Answer to the Q2:

```python

df['Salary']=df.groupby('Department')['Salary'].transform(lambda x: x.fillna(x.median()))
condition=df[(df['ExperienceYears']>12)&(df['ProjectCount']>5)]
total_salary=condition['Salary'].sum()
print(total_salary)
```

## Answer to the Q3

```python
filtered_condition=df[((df['Department']=='IT') | (df['Department']=='Finance')) &(df['Age'].between(30,45))&(df['PerformanceScore']>88)]
srt=filtered_condition.sort_values(by='Salary',ascending=False)
print(srt[['Name','Department','Salary','PerformanceScore']])


```

## Answer to the Q4:

```python
df['SalaryPerHour']=df['Salary']/(df['WorkHoursPerWeek']*4)
five=df.sort_values(by='SalaryPerHour',ascending=False).head(5)
print(five[['Name','SalaryPerHour']])



```
