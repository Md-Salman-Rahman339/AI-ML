Activation function adds non-linearity to the neural network.Neural network first of all calculate linear calculation of every neuron with this formula:z=wx+b
In real life maximum problem is non-linear like XOR problem.Activation function change the linear output into non-linear output with ReLU ,sigmoid ,Tanh.For these activation function:

1.Network can learn complex pattern
2.It will be easy for creating Non-linear decision boundary
3.We can solve complex problem
